```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sklearn 
from sklearn.linear_model import LinearRegression
```

```python
df1 = pd.read_json('emg1.json')
df2 = pd.read_json('emg2.json')
df3 = pd.read_json('emg3.json')
df1 = df1[100:-1]
df_combined = pd.concat([df1, df2, df3])
df_combined.reset_index(inplace=True)
```

```python
df_combined
```

```python
def plot_graph(X, Y):
    x = X.to_numpy()
    y = Y.to_numpy()
    m1, m2, b = np.polyfit(x, y, 2) #polynomial representation of degree 2
    plt.plot(x, y, 'o')
    plt.plot(x, m1 * x**2 + m2*x + b)
    plt.show()
```

```python
def get_mean(df1):
    df1_mean_freq = df1['meanFreq'].to_numpy()
    df1_voltageRMS = df1['voltageRMS'].to_numpy()
    df1_meanAmpy = df1['meanAmplitude'].to_numpy()
    mean_f = np.mean(df1_mean_freq)
    mean_v = np.mean(df1_voltageRMS)
    mean_a = np.mean(df1_meanAmpy)
    return mean_f, mean_v, mean_a
```

```python
print(get_mean(df1))
print(get_mean(df2))
print(get_mean(df3))
```

```python
plot_graph(df_combined['index'], df_combined['meanFreq'])
```

```python
plot_graph(df_combined['index'], df_combined['voltageRMS'])
```

```python
plot_graph(df_combined['index'], df_combined['meanAmplitude'])
```
